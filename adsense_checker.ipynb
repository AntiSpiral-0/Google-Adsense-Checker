# Going through the csv's sites

```python
import requests
import re
# OOP Method for searching sites that has google sense ads

class AdSenseChecker:
    def __init__(self) -> None:
        self.adsense_patterns = [
            re.compile(r'adsbygoogle.js'), # these are the patterns that google ad sense typically shows in the JS/HTML
            re.compile(r'googlesyndication.com'),
            re.compile(r'google_ad_client')
        ]
    
    def adsense_detecter(self, url):
        try:
            response = requests.get(url, timeout=10) # getting the URL
            if response.status_code == 200: # if the result of the response == 200 that means it's successfully retrieved
                content = response.text
                for pattern in self.adsense_patterns: # if the pattern is confirmed to be checked
                    if pattern.search(content): # returns true when its found
                        return True
            return False
        except requests.RequestException as e:
            print(f"Failed to retrieve {url} : {e}")
            return False

    def check_websites_from_csv(self, input_csv, output_csv):
        # reading the input CSV file
        df = pd.read_csv(input_csv)
        
        
        print(df.columns) # verify the column names again
        
        
        df['AdSense_Presence'] = df['Website'].apply(self.adsense_detecter)  # adding a new column for AdSense presence and applying detecter
        
        # writing the results to a new CSV file
        df.to_csv(output_csv, index=False)
        print("AdSense presence check completed. Results saved to", output_csv)


# Example usage
checker = AdSenseChecker()
checker.check_websites_from_csv(input_data, 'Results.csv')
```
